```python
from validate_email import validate_email
import pandas as pd
df = pd.read_csv(r"C:\Users\Adhaar Sharma\Jupyter Notebooks\FilteredEmails.csv")
# creating a list of dataframe columns

for i in range(len(df)) :
    email = str(df.iloc[i, 4])
    print(email)
    is_valid = validate_email(email, check_mx=True, verify=True)
    print(is_valid)
```

```python
df['isValid'] = df['Email'].apply(lambda x:validate_email(x, check_mx=True, verify=True))
```
